```python
import pandas as pd

path = '/content/northstar_data/northstar_dataset/'

customers = pd.read_csv(path + 'customers.csv')

print("Customers dataset loaded successfully!")
```

```python
# Fill missing numerical values using mean

customers['loyalty_score'].fillna(
    customers['loyalty_score'].mean(),
    inplace=True
)

# Fill missing categorical values using mode

customers['preferred_channel'].fillna(
    customers['preferred_channel'].mode()[0],
    inplace=True
)

print("Missing values handled successfully!")
```

```python
customers.isnull().sum()
```

```python
# Check duplicate records

customers.duplicated().sum()
```

```python
# Remove duplicates if any exist

customers.drop_duplicates(inplace=True)

print("Duplicate removal completed!")
```

```python
# Create customer loyalty categories

customers['loyalty_category'] = customers['loyalty_score'].apply(
    lambda x:
    'Low' if x < 40 else
    'Medium' if x < 70 else
    'High'
)

# Display results

customers[['customer_id',
           'loyalty_score',
           'loyalty_category']].head()
```

```python
# Create engagement categories

customers['engagement_level'] = customers['app_engagement_score'].apply(
    lambda x:
    'Low' if x < 40 else
    'Medium' if x < 70 else
    'High'
)

# Display results

customers[['customer_id',
           'app_engagement_score',
           'engagement_level']].head()
```

```python
import sqlite3

# Create database connection
conn = sqlite3.connect('northstar.db')

print("Database created successfully!")
```

```python
# Import customers table into database

customers.to_sql(
    'customers',
    conn,
    if_exists='replace',
    index=False
)

print("Customers table imported successfully!")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
# Average loyalty score by customer type

query = """
SELECT customer_type,
       AVG(loyalty_score) AS avg_loyalty_score
FROM customers
GROUP BY customer_type
"""

pd.read_sql(query, conn)
```

```python
# Average engagement score by customer type

query = """
SELECT customer_type,
       AVG(app_engagement_score) AS avg_engagement_score
FROM customers
GROUP BY customer_type
"""

pd.read_sql(query, conn)
```

```python
# Preferred communication channels count

query = """
SELECT preferred_channel,
       COUNT(*) AS total_customers
FROM customers
GROUP BY preferred_channel
ORDER BY total_customers DESC
"""

pd.read_sql(query, conn)
```

```python
# High loyalty customers

query = """
SELECT customer_id,
       customer_type,
       loyalty_score
FROM customers
WHERE loyalty_score >= 70
ORDER BY loyalty_score DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
# Query execution using indexed column

query = """
SELECT customer_type,
       AVG(loyalty_score) AS avg_loyalty
FROM customers
GROUP BY customer_type
"""

pd.read_sql(query, conn)
```
